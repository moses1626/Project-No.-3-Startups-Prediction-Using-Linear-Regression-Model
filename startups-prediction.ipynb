```python
# seperating inputs===> x   and output===> y

x=a.iloc[:,:-1]
y=a.iloc[:,-1]
x
```

```python
# from the above graphs, it is clear that R&D will affect more for profit coz there are a lot of points which passes through the best fit line

# converting 'State' into numerical value, so using encoding technique
# instead of using 'LabelEncoding' we are now using 'One_Hot_Encoding'
```

```python
from numpy.ma.core import remainder
# implementing One_Hot_Encoding
from sklearn.compose import make_column_transformer  #make_column_transformer means which column should be transformed
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
```

```python
# Implementing training and testing on the x and y

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)

x_train
```

```python
y_test
```

```python
# Model creation using LinearRegression

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
# Performance Evaluation

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('Mean Absolte Error is',mean_absolute_error(y_test,y_pred))
print('mean absolute percentage error is',mean_absolute_percentage_error(y_test,y_pred))
print('mean aquared error is',mean_squared_error(y_test,y_pred))
print('r2_score is',r2_score(y_test,y_pred))
mse=mean_squared_error(y_test,y_pred)
rsme=np.sqrt(mse)
print('root mean squared erorr is',rsme)
```

```python
# constant(intercept) and slope(coeffeicent)

print("the slope is",model.coef_)
print("the constant value is",model.intercept_)
# we need to find the slopes of each and every independent variables

list(zip(x,model.coef_))
```
